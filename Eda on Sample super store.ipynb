### Sales Vs profit

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
plt.plot(Profi)
plt.title('Total Profit')
plt.xlabel('Number of sales')
plt.ylabel('Profit')
plt.show()
```

```python
correla = sample_df.corr()
plt.figure(figsize=(15,8))
sns.heatmap(correla,annot = True )
plt.show()
```

### Most Used Ship Mode

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=75)
sns.countplot(sample_df['Ship Mode'])
plt.title('Highely Used Ship Mode')
plt.xlabel('Types of Ship Mode')
plt.ylabel('No Of Sales')
plt.show()
```

### We can observe that the `standard class` is the most used ship mode type 

### Heighest Sales Placed by a State

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=75)
sns.countplot(sample_df['State'])
plt.title('Highest order placed by a state')
plt.xlabel('States')
plt.ylabel('No Of Orders')
plt.show()
```

### Heighest Sales Placed by a Region

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=75)
sns.countplot(sample_df['Region'])
plt.title('Highest No Of order placed by region')
plt.xlabel('Region')
plt.ylabel('No Of Orders')
plt.show()
```

# Heighest sealing Products

```python
Sub_category = sample_df.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values(by='Sales',ascending=False).head(10)
```

```python
Sub_category
```

```python
plt.figure(figsize=(10,6))
plt.title('Top 10 Heighest sealing products')
plt.ylabel('Sales')
plt.xlabel('Products')
plt.bar(Sub_category['Sub-Category'],Sub_category['Sales']);
```

### Here we can observe that the `phones` and `chairs` are one of the heighest sealing products

# Most Profitable Products

```python
Heighest_profit=sample_df.groupby('Sub-Category')['Profit'].sum().reset_index().sort_values(by='Profit',ascending=False).head(10)
```

```python
Heighest_profit
```

```python
plt.figure(figsize=(10,6))
plt.title('Top 10 Heighest sealing products')
plt.ylabel('Sales')
plt.xlabel('Products')
plt.bar(Heighest_profit['Sub-Category'],Heighest_profit['Profit']);
```

### Here we observe the top 10 heighley profitable products.
